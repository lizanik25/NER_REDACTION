```python
import pandas as pd
import random
import re
import json
from faker import Faker

SEED = 42
random.seed(SEED)
Faker.seed(SEED)

fake_ru = Faker("ru_RU")
fake_en = Faker("en_US")


def shuffle_df(df):
    return df.sample(frac=1, random_state=SEED).reset_index(drop=True).copy()

def take_rows(df, start, n):
    return df.iloc[start:start+n].copy()

def normalize_spaces(text: str) -> str:
    text = re.sub(r"\s+", " ", text).strip()
    return text

def ensure_final_punct(text: str) -> str:
    text = text.strip()
    if not text:
        return text
    if text[-1] not in ".!?":
        text += "."
    return text

def gen_phone_core():
    code = random.choice(["495", "499", "812", "916", "925", "926", "903", "985"])
    a = random.randint(100, 999)
    b = random.randint(10, 99)
    c = random.randint(10, 99)

    fmt = random.choice(["A1", "A2", "A4", "A5"])

    if fmt == "A1":
        prefix = random.choice(["+7", "8", "7"])
        value = f"{prefix}{code}{a:03d}{b:02d}{c:02d}"

    elif fmt == "A2":
        prefix = random.choice(["+7", "8"])
        value = f"{prefix} {code} {a:03d} {b:02d} {c:02d}"

    elif fmt == "A4":
        prefix = random.choice(["+7", "8", ""])
        if prefix:
            value = f"{prefix} ({code}) {a:03d}-{b:02d}-{c:02d}"
        else:
            value = f"({code}) {a:03d}-{b:02d}-{c:02d}"

    else:
        prefix = random.choice(["+7", "8"])
        value = f"{prefix}-{code}-{a:03d}-{b:02d}-{c:02d}"

    return value, fmt

def gen_phone_rest():
    code = random.choice(["495", "499", "812", "916", "925", "926", "903", "985"])
    a = random.randint(100, 999)
    b = random.randint(10, 99)
    c = random.randint(10, 99)

    fmt = random.choice([
        "A3", "A6", "A7", "A8",
        "B1", "B2",
        "C1", "D", "E", "F", "G", "H", "I"
    ])

    if fmt == "A3":
        value = random.choice([
            f"+7 {code} {random.randint(10,99)} {random.randint(10,99)} {random.randint(100,999)}",
            f"8 {code} {random.randint(1000,9999)} {random.randint(100,999)}",
            f"+7 {code} {random.randint(100,999)} {random.randint(1,9)} {random.randint(100,999)}"
        ])

    elif fmt == "A6":
        prefix = random.choice(["+7", "8"])
        value = f"{prefix}.{code}.{a:03d}.{b:02d}.{c:02d}"

    elif fmt == "A7":
        value = random.choice([
            f"+7 ({code}) {a:03d} {b:02d}-{c:02d}",
            f"8 ({code}){a:03d} {b:02d} {c:02d}",
            f"+7-{code} {a:03d}-{b:02d} {c:02d}"
        ])

    elif fmt == "A8":
        prefix = random.choice(["+7", "8"])
        value = f"{prefix}({code}){a:03d}{b:02d}{c:02d}"

    elif fmt == "B1":
        value = random.choice([
            f"{code}{a:03d}{b:02d}{c:02d}",
            f"{code} {a:03d} {b:02d} {c:02d}"
        ])

    elif fmt == "B2":
        value = random.choice([
            f"({code}) {a:03d}-{b:02d}-{c:02d}",
            f"(495) {a:03d}-{b:02d}-{c:02d}"
        ])

    elif fmt == "C1":
        value = f"{a:03d}-{b:02d}-{c:02d}"

    elif fmt == "D":
        base = random.choice([
            f"+7 (495) {a:03d}-{b:02d}-{c:02d}",
            f"8-495-{a:03d}-{b:02d}-{c:02d}",
            f"+7495{a:03d}{b:02d}{c:02d}"
        ])
        ext = random.randint(10, 9999)
        suffix = random.choice([f"доб. {ext}", f"(вн. {ext})", f"ext {ext}", f"#{ext}"])
        value = f"{base} {suffix}"

    elif fmt == "E":
        prefix = random.choice([
            "тел.",
            "телефон:",
            "контактный номер:",
            "моб.:",
            "звонить по номеру",
            "связаться по"
        ])
        base, _ = gen_phone_core()
        value = f"{prefix} {base}"

    elif fmt == "F":
        base, _ = gen_phone_core()
        value = random.choice([
            f"тел.{base}",
            f"Иванов{base}",
            f"email@test.ru{base}",
            f"№{base}"
        ])

    elif fmt == "G":
        value = random.choice([
            f"тел. 8(495){a:03d}-{b:02d}-{c:02d}, 68, 69",
            f"тел.: {a:03d}-{b:02d}-{c:02d} / 68",
            f"тел.: {a:03d}-{b:02d}-{c:02d} и {a:03d}-{b:02d}-{c+1 if c < 99 else c:02d}"
        ])

    elif fmt == "H":
        value = "восемь девятьсот двадцать шесть сто двадцать три сорок пять шестьдесят семь"

    else:
        value = random.choice([
            f"+7({code}{a:03d}{b:02d}{c:02d}",
            f"8{code}-{a:03d}{b:02d}{c:02d}",
            f"8 {code}{a:03d} {b:02d} {c:02d}",
            f"+7 {code}{a:03d}{b:02d}{c:02d}"
        ])

    return value, fmt

def gen_phone(core=True):
    return gen_phone_core() if core else gen_phone_rest()


RUS_FIRST = ["ivan", "petr", "anna", "olga", "sergey", "alexey", "maria", "elena", "dmitry"]
RUS_LAST = ["ivanov", "petrov", "sidorov", "smirnov", "volkov", "nikitin", "popov", "fedorov"]

def gen_email_core():
    domains = ["mail.ru", "gmail.com", "yandex.ru", "bk.ru", "inbox.ru", "list.ru"]
    first = random.choice(RUS_FIRST)
    last = random.choice(RUS_LAST)
    number = random.randint(1, 2026)
    fmt = random.choice(["A1", "A2", "A3", "A4"])

    if fmt == "A1":
        local = random.choice([
            first,
            last,
            f"{first}{last}",
            f"user{number}"
        ])
    elif fmt == "A2":
        local = random.choice([
            f"{first}.{last}",
            "a.b.c",
            f"{first}.{random.choice(RUS_LAST)}"
        ])
    elif fmt == "A3":
        local = random.choice([
            f"{last}{number}",
            f"user{number}",
            f"{first}{random.randint(1980,2025)}"
        ])
    else:
        local = random.choice([
            f"{first}_{last}",
            f"{first}.{last}-{number}",
            f"super-man",
            f"{last}_{first}"
        ])

    return f"{local}@{random.choice(domains)}", fmt

def gen_email_rest():
    fmt = random.choice(["B", "C", "D", "E", "G", "H", "I", "J", "K", "M", "N", "O"])

    first = random.choice(RUS_FIRST)
    last = random.choice(RUS_LAST)
    domain = random.choice(["company.ru", "site.org", "bank.ru", "domain.ru", "service.org"])

    if fmt == "B":
        value = random.choice([
            f"{last}.{first}@mail.ru",
            f"{first[0]}.{last}@mail.ru",
            f"{first}.ivanovich.{last}@mail.ru",
            f"{last}_{first[0]}@mail.ru"
        ])

    elif fmt == "C":
        value = random.choice([
            "support@company.ru",
            "hr@company.ru",
            "admin@site.org",
            "info@bank.ru",
            "office@company.ru",
            "helpdesk@domain.ru"
        ])

    elif fmt == "D":
        value = f"{last}+urgent@{domain}"

    elif fmt == "E":
        value = random.choice([
            "иван@почта.рф",
            "директор@иванов.москва"
        ])

    elif fmt == "G":
        value = random.choice([
            "noreply@domain.ru",
            "mailer-daemon@google.com"
        ])

    elif fmt == "H":
        base, _ = gen_email_core()
        value = random.choice([
            f"{base}.",
            f"{base},",
            f"({base})",
            f"[{base}]"
        ])

    elif fmt == "I":
        value = random.choice([
            f"{last} @ mail . ru",
            f"{last}@ mail.ru",
            f"{last} @mail.ru"
        ])

    elif fmt == "J":
        value = random.choice([
            f"{last}@gmal.com",
            f"{last}@yandx.ru",
            f"ivannov@mail.ru",
            f"{last}@mail,ru"
        ])

    elif fmt == "K":
        base, _ = gen_email_core()
        value = random.choice([
            f"email:{base}",
            f"{base}тел",
            f"Иванов{base}"
        ])

    elif fmt == "M":
        value = random.choice([
            f"user@mail.company.ru",
            f"{last}@sub.domain.com"
        ])

    elif fmt == "N":
        value = random.choice([
            f"{last}@very-long-company-name.ru",
            f"user@my-super-service.org"
        ])

    else:
        value = random.choice([
            "IVANOV@MAIL.RU",
            "Ivanov@Gmail.Com"
        ])

    return value, fmt

def gen_email(core=True):
    return gen_email_core() if core else gen_email_rest()


RUS_CITIES = ["Москва", "Санкт-Петербург", "Казань", "Самара", "Краснодар", "Екатеринбург", "Сочи", "Балашиха"]
RUS_STREETS = ["Тверская", "Ленина", "Мясницкая", "Невский проспект", "Ново-Садовая", "Приморский проспект", "Калараш", "Сибирский тракт"]

def gen_address_core():

    city = random.choice(RUS_CITIES)
    street = random.choice(RUS_STREETS)
    house = random.randint(1, 120)
    flat = random.randint(1, 250)
    building = random.choice(["", f", стр. {random.randint(1,5)}", f", корп. {random.randint(1,4)}"])
    postal = random.randint(101000, 199999)

    fmt = random.choice(["A1", "A2", "A3", "B2"])

    if fmt == "A1":
        value = f"{postal}, г. {city}, ул. {street}, д. {house}{building}, кв. {flat}"
    elif fmt == "A2":
        value = f"г. {city}, ул. {street}, дом {house}, квартира {flat}"
    elif fmt == "A3":
        value = f"г. {city}, {street}, д. {house}"
    else:
        value = random.choice([
            f"{street} {house}-{flat}",
            f"{street} {house} стр {random.randint(1,5)}"
        ])

    return value, fmt

def gen_address_rest():

    city = random.choice(RUS_CITIES)
    street = random.choice(RUS_STREETS)
    house = random.randint(1, 120)
    flat = random.randint(1, 250)
    postal = random.randint(101000, 199999)

    fmt = random.choice(["C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O", "P", "Q"])

    if fmt == "C":
        value = f"МО, г. {city}, пр-кт {street}, д. {house}, кв. {flat}"

    elif fmt == "D":
        value = f"Калужская обл., Боровский р-н, д. Митяево, д. {house}"

    elif fmt == "E":
        value = random.choice([
            f"г. Сочи, пос. Лазаревское, ул. Калараш, гостевой дом 'Светлана'",
            f"МО, дер. Ивановка, 3-й дом от магазина",
            f"рядом с ТЦ 'Глобус'",
            f"напротив школы №5"
        ])

    elif fmt == "F":
        value = random.choice([
            "км 104 трассы М-4 Дон, владение 2",
            "г. Екатеринбург, Сибирский тракт, 5 км, литер А"
        ])

    elif fmt == "G":
        value = "г. Краснодар, Прикубанский округ, ГПЗ-24, уч. 14"

    elif fmt == "H":
        value = "в/ч 12345, общежитие №3"

    elif fmt == "I":
        value = random.choice([
            "123 Main Street, Apt 4B, New York, NY, 10001, USA",
            "ul. Lenina, d. 15, kv. 32, Moscow, Russia"
        ])

    elif fmt == "J":
        value = random.choice([
            f"{postal}, г. {city}, ул. {street}, д. {house}",
            f"г. {city}, {postal}, ул. {street}, д. {house}"
        ])

    elif fmt == "K":
        value = random.choice([
            f"{city} ул {street} д {house} кв {flat}",
            f"г {city} {street} {house}"
        ])

    elif fmt == "L":
        value = random.choice([
            f"адрес:г.{city},ул.{street},д.{house}",
            f"{city},ул.{street},д.{house}кв.{flat}"
        ])

    elif fmt == "M":
        value = random.choice([
            f"ул {street}",
            f"д {house}",
            f"кв {flat}"
        ])

    elif fmt == "N":
        value = random.choice([
            f"д. {house}А",
            f"кв. {flat}Б",
            "лит. А"
        ])

    elif fmt == "O":
        value = random.choice([
            f"адреса: Москва, ул. Тверская, д. 1; Санкт-Петербург, Невский пр., д. 5"
        ])

    elif fmt == "P":
        value = random.choice([
            "этаж 3",
            "5 этаж"
        ])

    else:
        value = random.choice([
            "корп. 2",
            "стр. 1",
            "вл. 5"
        ])

    return value, fmt

def gen_address(core=True):
    return gen_address_core() if core else gen_address_rest()


def gen_id_core():

    fmt = random.choice(["A", "B", "C"])

    if fmt == "A":
        value = random.choice([
            f"ID: {random.randint(10000, 99999)}",
            f"№{random.randint(10000, 99999)}",
            f"id={random.randint(10000, 99999)}",
            f"user_id:{random.randint(10000, 99999)}"
        ])

    elif fmt == "B":
        letters = "".join(random.choices("АБВГДЕЖЗИКЛМНОПРСТУФХ", k=2))
        value = random.choice([
            f"договор №{letters}-{random.randint(2024,2026)}/{random.randint(1,12):02d}-{random.randint(1,99):02d}",
            f"контракт {random.randint(10,99)}-{random.randint(100,999)}-{random.choice(['А','Б','В'])}",
            f"заявка №{random.randint(2024,2026)}-{random.randint(1,99999):05d}"
        ])

    else:
        snils = f"{random.randint(100,999)}-{random.randint(100,999)}-{random.randint(100,999)} {random.randint(10,99)}"
        inn = "".join(str(random.randint(0, 9)) for _ in range(12))
        value = random.choice([
            f"ИНН {inn}",
            inn,
            snils,
            snils.replace("-", "").replace(" ", "")
        ])

    return value, fmt

def gen_id_rest():

    fmt = random.choice(["D", "E", "F", "G", "H"])

    if fmt == "D":
        acct = "40817810" + "".join(str(random.randint(0,9)) for _ in range(12))
        value = random.choice([
            acct,
            f"р/с {acct}"
        ])

    elif fmt == "E":
        value = random.choice([
            f"ИНН:{''.join(str(random.randint(0, 9)) for _ in range(12))}",
            f"user_id:{random.randint(10000, 99999)}",
            f"№{random.randint(10000,99999)}"
        ])

    elif fmt == "F":
        value = random.choice([
            f"номер заявки {random.randint(100,99999)}",
            f"ID клиента {random.randint(10000,99999)}",
            f"договор {random.randint(100,999)}-{random.randint(10,99)}",
            f"счет {random.randint(100000,999999)}"
        ])

    elif fmt == "G":
        base = random.randint(100, 999)
        value = random.choice([
            f"№{base}, №{base+1}",
            f"заявки: {base}, {base+1}, {base+2}"
        ])

    else:
        value = random.choice([
            str(random.randint(900000000, 999999999)),
            "40817810000001234567"
        ])

    return value, fmt

def gen_id(core=True):
    return gen_id_core() if core else gen_id_rest()



def make_core_rest_pool(total_count, core_ratio=0.8):
    core_count = int(round(total_count * core_ratio))
    rest_count = total_count - core_count
    pool = [True] * core_count + [False] * rest_count
    random.shuffle(pool)
    return pool

PHONE_TOTAL = 450
EMAIL_TOTAL = 400
ADDRESS_TOTAL = 357
ID_TOTAL = 340

phone_core_pool = make_core_rest_pool(PHONE_TOTAL, 0.8)
email_core_pool = make_core_rest_pool(EMAIL_TOTAL, 0.8)
address_core_pool = make_core_rest_pool(ADDRESS_TOTAL, 0.8)
id_core_pool = make_core_rest_pool(ID_TOTAL, 0.8)

def pop_entity_value(entity_type):
    if entity_type == "PHONE":
        core = phone_core_pool.pop()
        value, fmt = gen_phone(core=core)
    elif entity_type == "EMAIL":
        core = email_core_pool.pop()
        value, fmt = gen_email(core=core)
    elif entity_type == "ADDRESS":
        core = address_core_pool.pop()
        value, fmt = gen_address(core=core)
    elif entity_type == "ID":
        core = id_core_pool.pop()
        value, fmt = gen_id(core=core)
    else:
        raise ValueError(f"Unknown entity type: {entity_type}")

    return {
        "type": entity_type,
        "value": value,
        "format": fmt,
        "is_core_format": core
    }


ENTITY_ORDER = {
    "PHONE": 0,
    "EMAIL": 1,
    "ADDRESS": 2,
    "ID": 3,
}

DOC_TYPES_BY_N = {
    1: ["application", "service_request", "questionnaire", "explanatory_note", "hr_form"],
    2: ["application", "service_request", "questionnaire", "explanatory_note", "contract_fragment"],
    3: ["application", "service_request", "questionnaire", "contract_fragment", "court_material"],
    4: ["application", "questionnaire", "contract_fragment", "court_material"],
}

NEWSY_SOURCES = {"nerus", "nerel", "factrueval2016", "wikipedia", "winer", "mokoron"}

SYNTH_SOURCES = {"synthetic"}

BODY_INTROS = {
    "application": [
        "В заявлении указано следующее: {body}",
        "В тексте обращения указано: {body}",
        "Заявитель сообщает следующее: {body}",
        "В обращении изложены следующие обстоятельства: {body}",
        "В представленных пояснениях указано: {body}",
    ],
    "service_request": [
        "В описании обращения указано: {body}",
        "Суть запроса сформулирована следующим образом: {body}",
        "В карточке обращения указано: {body}",
        "По информации из обращения: {body}",
        "В тексте запроса отмечено: {body}",
    ],
    "questionnaire": [
        "В разделе «Дополнительные сведения» указано: {body}",
        "В анкете приведены следующие сведения: {body}",
        "В комментарии к анкете указано: {body}",
        "В пояснениях к анкете отмечено: {body}",
    ],
    "hr_form": [
        "В личной карточке указано: {body}",
        "В кадровых материалах отмечено: {body}",
        "В представленных сведениях указано: {body}",
    ],
    "contract_fragment": [
        "В приложении к документам указано: {body}",
        "В представленных документах содержатся сведения: {body}",
        "В договорных материалах указано: {body}",
    ],
    "court_material": [
        "В материалах дела указано: {body}",
        "Согласно представленным материалам: {body}",
        "В пояснениях по делу указано: {body}",
    ],
    "explanatory_note": [
        "В объяснении указано: {body}",
        "В тексте пояснения отмечено: {body}",
        "Из объяснительной следует: {body}",
        "В служебной записке указано: {body}",
    ],
}

DOC_TEMPLATES = {
    "application": [
        "Заявление\n\n{body_block}\n\n{fields_block}",
        "Обращение\n\n{body_block}\n\n{fields_block}",
        "Письменное обращение\n\n{body_block}\n\n{fields_block}",
    ],
    "service_request": [
        "Карточка обращения\n\n{body_block}\n\n{fields_block}",
        "Сведения по обращению\n\n{body_block}\n\n{fields_block}",
        "Заявка на обработку\n\n{body_block}\n\n{fields_block}",
    ],
    "questionnaire": [
        "Анкета\n\n{body_block}\n\n{fields_block}",
        "Анкетные сведения\n\n{body_block}\n\n{fields_block}",
        "Форма персональных данных\n\n{body_block}\n\n{fields_block}",
    ],
    "hr_form": [
        "Личные данные\n\n{body_block}\n\n{fields_block}",
        "Карточка сотрудника\n\n{body_block}\n\n{fields_block}",
        "Кадровая анкета\n\n{body_block}\n\n{fields_block}",
    ],
    "contract_fragment": [
        "Сведения по договору\n\n{body_block}\n\n{fields_block}",
        "Реквизиты стороны\n\n{body_block}\n\n{fields_block}",
        "Фрагмент договорных материалов\n\n{body_block}\n\n{fields_block}",
    ],
    "court_material": [
        "Материалы дела\n\n{body_block}\n\n{fields_block}",
        "Процессуальные материалы\n\n{body_block}\n\n{fields_block}",
        "Сведения по делу\n\n{body_block}\n\n{fields_block}",
    ],
    "explanatory_note": [
        "Пояснение\n\n{body_block}\n\n{fields_block}",
        "Объяснительная записка\n\n{body_block}\n\n{fields_block}",
        "Служебная записка\n\n{body_block}\n\n{fields_block}",
    ],
}

SECTION_HEADERS = {
    "application": [
        "Контактные данные заявителя:",
        "Дополнительные сведения:",
        "Сведения для обратной связи:",
    ],
    "service_request": [
        "Контактные данные:",
        "Сведения для обратной связи:",
        "Указанные реквизиты:",
    ],
    "questionnaire": [
        "Контактные сведения:",
        "Персональные данные:",
        "Указанные данные:",
    ],
    "hr_form": [
        "Контактные сведения:",
        "Реквизиты сотрудника:",
        "Дополнительные данные:",
    ],
    "contract_fragment": [
        "Реквизиты стороны:",
        "Контактные данные:",
        "Идентификационные сведения:",
    ],
    "court_material": [
        "Данные участника:",
        "Сведения для связи:",
        "Дополнительные реквизиты:",
    ],
    "explanatory_note": [
        "Контактные данные:",
        "Дополнительные сведения:",
        "Указанные реквизиты:",
    ],
}

ENTITY_LABELS = {
    "PHONE": [
        "Телефон",
        "Контактный телефон",
        "Номер телефона",
        "Тел.",
        "Мобильный телефон",
    ],
    "EMAIL": [
        "Электронная почта",
        "Email",
        "Адрес электронной почты",
        "Почта для связи",
    ],
    "ADDRESS": [
        "Адрес регистрации",
        "Адрес проживания",
        "Адрес для корреспонденции",
        "Фактический адрес",
        "Место проживания",
    ],
    "ID": [
        "Идентификатор",
        "Номер документа",
        "Регистрационный номер",
        "Номер обращения",
        "ID клиента",
    ],
}

INLINE_CONNECTORS = {
    "application": [
        "Для обратной связи заявитель указал",
        "В заявлении дополнительно указаны",
        "В обращении также приведены",
    ],
    "service_request": [
        "Для обратной связи указаны",
        "В обращении дополнительно приведены",
        "В карточке обращения указаны",
    ],
    "questionnaire": [
        "В анкете указаны",
        "Дополнительно в анкете приведены",
        "В анкете для связи указаны",
    ],
    "hr_form": [
        "В личной карточке указаны",
        "В кадровых материалах дополнительно указаны",
    ],
    "contract_fragment": [
        "В реквизитах стороны указаны",
        "В договорных материалах дополнительно приведены",
    ],
    "court_material": [
        "В материалах дела дополнительно указаны",
        "В сведениях по делу приведены",
    ],
    "explanatory_note": [
        "В пояснении дополнительно указаны",
        "В служебной записке приведены",
    ],
}

INLINE_ENTITY_PATTERNS = {
    "PHONE": [
        "телефон {value}",
        "контактный телефон {value}",
        "номер телефона {value}",
    ],
    "EMAIL": [
        "электронная почта {value}",
        "email {value}",
        "адрес электронной почты {value}",
    ],
    "ADDRESS": [
        "адрес проживания {value}",
        "адрес для корреспонденции {value}",
        "фактический адрес {value}",
    ],
    "ID": [
        "номер обращения {value}",
        "идентификатор {value}",
        "регистрационный номер {value}",
        "номер документа {value}",
    ],
}

def choose_doc_type(source, n_added_types):
    source = str(source).lower()
    pool = DOC_TYPES_BY_N[n_added_types]

    if source in SYNTH_SOURCES:
        weights = {
            "application": 4,
            "questionnaire": 4,
            "service_request": 3,
            "hr_form": 2,
            "explanatory_note": 2,
            "contract_fragment": 1,
            "court_material": 0,
        }
    elif source in NEWSY_SOURCES:
        weights = {
            "application": 4,
            "service_request": 4,
            "explanatory_note": 3,
            "questionnaire": 2,
            "contract_fragment": 1,
            "court_material": 1 if n_added_types >= 3 else 0,
            "hr_form": 0,
        }
    else:
        weights = {doc: 1 for doc in pool}

    weighted_pool = []
    for doc in pool:
        weighted_pool.extend([doc] * weights.get(doc, 1))

    return random.choice(weighted_pool if weighted_pool else pool)

def cleanup_real_sentence(text):
    text = str(text)
    text = re.sub(r"\s+", " ", text).strip()
    text = re.sub(r"\s+([,.;:!?])", r"\1", text)
    text = text.replace(" .", ".")
    text = text.replace(" ,", ",")
    text = text.replace(" :", ":")
    text = text.replace(" ;", ";")
    text = text.replace("« ", "«").replace(" »", "»")
    return ensure_final_punct(text)

def make_body_block(real_text, doc_type):
    real_text = cleanup_real_sentence(real_text)

    intro = random.choice(BODY_INTROS[doc_type])
    return intro.format(body=real_text)

def build_structured_fields_block(entities, doc_type):
    entities = sorted(entities, key=lambda x: ENTITY_ORDER[x["type"]])
    lines = [random.choice(SECTION_HEADERS[doc_type])]

    for ent in entities:
        label = random.choice(ENTITY_LABELS[ent["type"]])
        lines.append(f"{label}: {ent['value']}")

    return "\n".join(lines)

def build_inline_fields_block(entities, doc_type):
    entities = sorted(entities, key=lambda x: ENTITY_ORDER[x["type"]])

    parts = []
    for ent in entities:
        pattern = random.choice(INLINE_ENTITY_PATTERNS[ent["type"]])
        parts.append(pattern.format(value=ent["value"]))

    connector = random.choice(INLINE_CONNECTORS[doc_type])

    if len(parts) == 1:
        return f"{connector} {parts[0]}."
    elif len(parts) == 2:
        return f"{connector} {parts[0]} и {parts[1]}."
    else:
        return f"{connector} " + ", ".join(parts[:-1]) + f" и {parts[-1]}."

def choose_fields_mode(doc_type, source, n_added_types):
    source = str(source).lower()

    if doc_type in {"questionnaire", "hr_form", "contract_fragment", "court_material"}:
        return "structured"

    if source in NEWSY_SOURCES:
        return random.choice(["inline", "structured", "inline"])

    if source in SYNTH_SOURCES:
        return random.choice(["structured", "inline", "structured"])

    return random.choice(["structured", "inline"])

def build_fields_block(entities, doc_type, source):
    mode = choose_fields_mode(doc_type, source, len(entities))
    if mode == "structured":
        return build_structured_fields_block(entities, doc_type)
    return build_inline_fields_block(entities, doc_type)

def preserve_document_formatting(text):
    lines = text.split("\n")
    lines = [re.sub(r"\s+", " ", line).strip() for line in lines]
    cleaned = []
    prev_empty = False
    for line in lines:
        is_empty = (line == "")
        if is_empty and prev_empty:
            continue
        cleaned.append(line)
        prev_empty = is_empty

    text = "\n".join(cleaned).strip()
    return text

def render_document_text(real_text, combo, source):
    combo = tuple(sorted(combo))
    entities = [pop_entity_value(t) for t in combo]
    doc_type = choose_doc_type(source, len(combo))

    body_block = make_body_block(real_text, doc_type)
    fields_block = build_fields_block(entities, doc_type, source)

    template = random.choice(DOC_TEMPLATES[doc_type])
    mixed_text = template.format(
        body_block=body_block,
        fields_block=fields_block
    )

    mixed_text = preserve_document_formatting(mixed_text)
    synthetic_part = fields_block

    return mixed_text, synthetic_part, entities, doc_type

plus1_combos = (
    [("PHONE",)] * 150 +
    [("EMAIL",)] * 130 +
    [("ADDRESS",)] * 130 +
    [("ID",)] * 113
)
assert len(plus1_combos) == 523

plus2_combos = (
    [("PHONE", "EMAIL")] * 70 +
    [("PHONE", "ID")] * 55 +
    [("PHONE", "ADDRESS")] * 55 +
    [("ID", "EMAIL")] * 40 +
    [("ID", "ADDRESS")] * 25 +
    [("ADDRESS", "EMAIL")] * 40
)
assert len(plus2_combos) == 285

plus3_combos = (
    [("PHONE", "EMAIL", "ID")] * 35 +
    [("PHONE", "EMAIL", "ADDRESS")] * 35 +
    [("PHONE", "ADDRESS", "ID")] * 22 +
    [("ID", "EMAIL", "ADDRESS")] * 22
)
assert len(plus3_combos) == 114

plus4_combos = [("PHONE", "EMAIL", "ADDRESS", "ID")] * 28
assert len(plus4_combos) == 28

random.shuffle(plus1_combos)
random.shuffle(plus2_combos)
random.shuffle(plus3_combos)
random.shuffle(plus4_combos)


df1 = shuffle_df(df_1_sampled_522)
df2 = shuffle_df(df_2_sampled_332)
df3 = shuffle_df(df_3_4_5_sampled_96)


df1_p1 = take_rows(df1, 0, 365)
df1_p2 = take_rows(df1, 365, 131)
df1_p3 = take_rows(df1, 496, 26)

df2_p1 = take_rows(df2, 0, 150)
df2_p2 = take_rows(df2, 150, 99)
df2_p3 = take_rows(df2, 249, 60)
df2_p4 = take_rows(df2, 309, 23)

df3_p1 = take_rows(df3, 0, 8)
df3_p2 = take_rows(df3, 8, 55)
df3_p3 = take_rows(df3, 63, 28)
df3_p4 = take_rows(df3, 91, 5)

assert len(df1_p1) + len(df2_p1) + len(df3_p1) == 523
assert len(df1_p2) + len(df2_p2) + len(df3_p2) == 285
assert len(df1_p3) + len(df2_p3) + len(df3_p3) == 114
assert len(df2_p4) + len(df3_p4) == 28

df1_p1["person_bucket"] = "1_person"
df1_p2["person_bucket"] = "1_person"
df1_p3["person_bucket"] = "1_person"

df2_p1["person_bucket"] = "2_person"
df2_p2["person_bucket"] = "2_person"
df2_p3["person_bucket"] = "2_person"
df2_p4["person_bucket"] = "2_person"

df3_p1["person_bucket"] = "3plus_person"
df3_p2["person_bucket"] = "3plus_person"
df3_p3["person_bucket"] = "3plus_person"
df3_p4["person_bucket"] = "3plus_person"

bucket_plus1_df = pd.concat([df1_p1, df2_p1, df3_p1], ignore_index=True)
bucket_plus2_df = pd.concat([df1_p2, df2_p2, df3_p2], ignore_index=True)
bucket_plus3_df = pd.concat([df1_p3, df2_p3, df3_p3], ignore_index=True)
bucket_plus4_df = pd.concat([df2_p4, df3_p4], ignore_index=True)

bucket_plus1_df = shuffle_df(bucket_plus1_df)
bucket_plus2_df = shuffle_df(bucket_plus2_df)
bucket_plus3_df = shuffle_df(bucket_plus3_df)
bucket_plus4_df = shuffle_df(bucket_plus4_df)



def process_bucket(df_bucket, combos, mix_bucket_label):
    result_rows = []

    assert len(df_bucket) == len(combos), f"{mix_bucket_label}: mismatch rows vs combos"

    for idx, combo in enumerate(combos):
        row = df_bucket.iloc[idx].copy()

        mixed_text, synthetic_part, added_entities, doc_type = render_document_text(
            real_text=row["sentence"],
            combo=combo,
            source=row.get("source", "")
        )

        row["mixed_text"] = mixed_text
        row["synthetic_part"] = synthetic_part
        row["added_types"] = ",".join(sorted(combo))
        row["added_entities_json"] = json.dumps(added_entities, ensure_ascii=False)
        row["mix_bucket"] = mix_bucket_label
        row["n_added_types"] = len(combo)
        row["doc_type"] = doc_type

        result_rows.append(row)

    return pd.DataFrame(result_rows)

mixed_plus1 = process_bucket(bucket_plus1_df, plus1_combos, "person_plus_1_type")
mixed_plus2 = process_bucket(bucket_plus2_df, plus2_combos, "person_plus_2_types")
mixed_plus3 = process_bucket(bucket_plus3_df, plus3_combos, "person_plus_3_types")
mixed_plus4 = process_bucket(bucket_plus4_df, plus4_combos, "person_plus_4_types")

df_mixed_950 = pd.concat(
    [mixed_plus1, mixed_plus2, mixed_plus3, mixed_plus4],
    ignore_index=True
)

df_mixed_950 = shuffle_df(df_mixed_950)


assert len(df_mixed_950) == 950

print("Размер итогового mixed-датасета:", df_mixed_950.shape)
print(df_mixed_950["mix_bucket"].value_counts())

print(df_mixed_950["person_bucket"].value_counts())

print(df_mixed_950[[
    "source", "sentence", "mixed_text", "person_bucket", "mix_bucket", "added_types"
]].head())

```

```python
import pandas as pd
import re


df_4_plus_final_21 = pd.read_csv("/content/df_4_plus_final_21 - df_4_plus_clean.csv (1).csv", encoding="utf-8")
df_3_final_153 = pd.read_csv("/content/df_3_final_153 - df_3_clean.csv (1).csv", encoding="utf-8")
df_1_final_639 = pd.read_csv("/content/df_1_final_639 - Лист1.csv", encoding="utf-8")
df_2_final_431 = pd.read_csv("/content/df_2_final_431 - df_2_clean_no_empty.csv (1).csv", encoding="utf-8")
df_mixed_947_final = pd.read_csv("/content/df_mixed_947_final - df_mixed_950_final.csv.csv", encoding="utf-8")


def normalize_sentence(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


df_all_person = pd.concat(
    [
        df_4_plus_final_21,
        df_3_final_153,
        df_1_final_639,
        df_2_final_431
    ],
    ignore_index=True,
    sort=False
).copy()


df_all_person["sentence_norm"] = df_all_person["sentence"].apply(normalize_sentence)
df_mixed_947_final["sentence_norm"] = df_mixed_947_final["sentence"].apply(normalize_sentence)


used_sentences = set(df_mixed_947_final["sentence_norm"])

df_remaining = df_all_person[
    ~df_all_person["sentence_norm"].isin(used_sentences)
].copy()


df_remaining = df_remaining.drop_duplicates(subset=["sentence_norm"]).reset_index(drop=True)

n_sample = 298


df_remaining_sample_298 = df_remaining.sample(
    n=n_sample,
    random_state=42
).copy()

df_remaining_sample_298 = df_remaining_sample_298.reset_index(drop=True)

print("Размер итоговой выборки:", df_remaining_sample_298.shape)


df_remaining_sample_298 = df_remaining_sample_298.drop(columns=["sentence_norm"])

```

```python
df_remaining_sample_298.to_csv(
    "df_remaining_sample_298.csv",
    index=False,
    encoding="utf-8-sig"
)
```

Сбор датасета без : No_PD и Tricky.

```python
import pandas as pd
import json
import ast


df_person_sample_298 = pd.read_csv("/content/df_person_sample_298 - df_remaining_sample_298.csv.csv", encoding="utf-8")
df_mixed_947_final = pd.read_csv("/content/df_mixed_947_final - df_mixed_950_final.csv (1).csv", encoding="utf-8")
synthetic_final_299_entities_v8 = pd.read_csv("/content/synthetic_final_299_entities_v8 - synthetic_300_entities_v8_balanced_counts.csv.csv", encoding="utf-8")


def safe_parse_list(value):
    if pd.isna(value):
        return []
    if isinstance(value, list):
        return value
    if isinstance(value, str):
        value = value.strip()
        if value == "":
            return []
        try:
            parsed = ast.literal_eval(value)
            if isinstance(parsed, list):
                return parsed
            return [parsed]
        except Exception:
            return [value]
    return [value]

def parse_added_entities_json(value):
    if pd.isna(value):
        return []
    if isinstance(value, list):
        data = value
    else:
        try:
            data = json.loads(value)
        except Exception:
            try:
                data = ast.literal_eval(value)
            except Exception:
                return []

    result = []
    if isinstance(data, list):
        for item in data:
            if isinstance(item, dict) and "value" in item:
                result.append(item["value"])
            else:
                result.append(item)
    return result


df_person_final = df_person_sample_298.copy()

df_person_final["source"] = df_person_final["source"]
df_person_final["text"] = df_person_final["sentence"]
df_person_final["n_entities"] = df_person_final["n_entities"]
df_person_final["entity_values"] = df_person_final["persons_raw"]

df_person_final = df_person_final[["source", "text", "n_entities", "entity_values"]].copy()


df_synth_final = synthetic_final_299_entities_v8.copy()

df_synth_final["source"] = "synthetic"
df_synth_final["text"] = df_synth_final["text"]
df_synth_final["n_entities"] = df_synth_final["n_entities_total"]
df_synth_final["entity_values"] = df_synth_final["entity_values"]

df_synth_final = df_synth_final[["source", "text", "n_entities", "entity_values"]].copy()


df_mixed_final = df_mixed_947_final.copy()

df_mixed_final["source"] = df_mixed_final["source"].astype(str) + " + synthetic"

df_mixed_final["text"] = df_mixed_final["mixed_text"]

df_mixed_final["n_entities"] = (
    pd.to_numeric(df_mixed_final["n_entities"], errors="coerce").fillna(0)
    + pd.to_numeric(df_mixed_final["n_added_types"], errors="coerce").fillna(0)
)

mixed_entity_values = []

for _, row in df_mixed_final.iterrows():
    persons = safe_parse_list(row.get("persons_raw"))
    added = parse_added_entities_json(row.get("added_entities_json"))
    combined = persons + added
    mixed_entity_values.append(str(combined))

df_mixed_final["entity_values"] = mixed_entity_values

df_mixed_final = df_mixed_final[["source", "text", "n_entities", "entity_values"]].copy()


df_final_combined = pd.concat(
    [df_person_final, df_mixed_final, df_synth_final],
    ignore_index=True
)


df_final_combined["text"] = df_final_combined["text"].astype(str).str.strip()
df_final_combined = df_final_combined[df_final_combined["text"] != ""].reset_index(drop=True)

df_final_combined["n_entities"] = pd.to_numeric(
    df_final_combined["n_entities"], errors="coerce"
).fillna(0).astype(int)


print("Размер итогового датасета:", df_final_combined.shape)
print(df_final_combined.head())

print(df_final_combined["source"].value_counts().head(20))

```

```python
df_final_combined["text"] = df_final_combined["text"].astype(str).str.strip()

df_final_combined = df_final_combined[
    (df_final_combined["text"].notna()) &
    (df_final_combined["text"] != "") &
    (df_final_combined["text"] != "nan")
].copy()
```

```python
df_final_combined.to_csv(
    "final_dataset_for_annotation.csv",
    index=False,
    encoding="utf-8-sig"
)

```
